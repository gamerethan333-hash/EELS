## Align ZLP

For atomic resolution EELS where each pixel is with scarce signal, bin the several pixels as 1 superpixel as reference to align the ZLP.

```python
import numpy as np
from scipy.ndimage import zoom

# 1. Bin by [2, 25] - must be perfect divisors for original shape.

llsi_binned = llsi.rebin(scale=[2, 25, 1])

# 2. Estimate ZLP centers (This creates the small drift map in RAM)
zlpc = llsi_binned.estimate_zero_loss_peak_centre()

# 3. Use scipy.zoom to restore the map to original 926x250
# We calculate the exact zoom factors needed to match your grid
zoom_x = 926 / zlpc.data.shape[1]
zoom_y = 250 / zlpc.data.shape[0]
shifts_array = zoom(zlpc.data, (zoom_y, zoom_x), order=1)

# 4. Apply to original signals with num_workers=1 for stability
# This keeps the kernel alive while fixing the drift
llsi.shift1D(shifts_array, num_workers=1)
hlsi.shift1D(shifts_array, num_workers=1)
```

## Spectrum Plotting

### Select the ROI

```python
# Plot ADF and attach interactive ROI

adf.plot(label="ADF Image")
plt.gca().set_title("ADF Image")
roi = hs.roi.RectangularROI()
roi_view = roi.interactive(adf, color="red")  # keep roi_view alive
```

### Assign ROI to the SI

```python
cropllsi = roi(llsi) #Crop original SI with roi
crophlsi = roi(hlsi)
```

## Low-loss Spectrum

```python
specll = cropllsi.sum(axis=('x', 'y')) # Create ll Spec based on cropped SI
specll.compute()
specll.plot()
plt.gca().set_title("Low-Loss EELS Spectrum")
```

### LL Elemental Analysis

### Element Search -- Low-loss

```python
specll.plot()           # make sure a plot exists and has valid limits
specll.edges_at_energy()
```

### Find element near certain Energy

```python
specll.print_edges_near_energy(401, width=20)
```

### Low-loss Spectrum with Elemant marks

```python
specll.plot(plot_edges=('Li','O'), only_edges=('Major', 'Minor'))
plt.gca().set_title("Low-Loss EELS Spectrum")
```

### Background Removal Low-loss

```python
specll_bgr = specll.deepcopy()
specll_bgr.plot(plot_edges=('Li','O'), only_edges=('Major', 'Minor'))
specll_bgr.remove_background()
plt.gca().set_title("Low-Loss EELS Spectrum - Background Subtracted")
```

### Thickness Estimation

```python
threshold = specll.estimate_elastic_scattering_threshold(window=10.0, tol=None, window_length=5, polynomial_order=3, start=1.0)
```

```python
thkns = specll.estimate_thickness(threshold= threshold, zlp=None, density=None, mean_free_path=None)
thkns_value = thkns.data.item()
print("t/lambda =", thkns_value)
```
